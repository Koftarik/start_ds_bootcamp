## Задание 11 (бонус)

Есть ли в данных другие страны, население которых также похожим образом ведет себя на поведение численности населения в Болгарии? Среди этих стран есть ли та, корреляция населения которой с населением Болгарии больше найденной в задании 7?

Перечислите в комментариях найденные страны, а также укажите страну-лидер по корреляции с населением Болгарии.

```python
countries = []

korr_koefs = []

# По каждой стране проверяем, что имел место максимум

for country in df['Country Name'].unique():

    population = df[df['Country Name'] == country].values[0][1:]

    mean = np.mean(df[df['Country Name'] == country].values[0][1:])

    # Если максимум был, то добавляем страну в список

    if population[0] < np.max(population) and population[-1] < np.max(population):

        countries.append(country)

        # вычисляем коэффициент корреляции Пирсона и добавляем его в список

        korr_koef = (sum((mean - population) * (bulg_pop.mean() - bulg_pop))

            / (sum((mean - population) ** 2) * sum((bulg_pop.mean() - bulg_pop) ** 2))** 0.5)

        korr_koefs.append(korr_koef)

# создадим датафрейм с двумя колонками

korr_df = pd.DataFrame(countries, columns=['country'])

korr_df['korr_koef'] = korr_koefs

# Отсортируем таблицу по убыванию коэффициента Пирсона и выведем первые 10 строк

# без Болгарии

korr_df[korr_df['country'] != 'Bulgaria'].sort_values(by=['korr_koef'], ascending=False).head(10)
```

```python
import seaborn as sns
# Отсортируем по убыванию коэффициента корреляции
# Отсортируем DataFrame по убыванию коэффициента корреляции
corr_df_sorted = korr_df.sort_values(by='korr_koef', ascending=False)

# Выберем первые 10 строк для отображения
top_corr_df = corr_df_sorted.head(10)

# Создадим матрицу корреляции, установив 'country' в качестве индекса,
# затем сортируем по 'korr_koef' для сохранения порядка сортировки
corr_matrix = top_corr_df.set_index('country')['korr_koef'].to_frame()  # Преобразуем в DataFrame

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, cmap='coolwarm', annot=True, cbar_kws={'label': 'Коэффициент корреляции Пирсона'})
plt.title('Тепловая карта корреляции населения стран с населением Болгарии')
plt.show()
```
